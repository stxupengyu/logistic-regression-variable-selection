```python
data1.columns
```

```python
data[['k']] = data['k'].apply(lambda x : 'a' if x==1 else 'b')
datakl=pd.get_dummies(data[['k','l']])
```

```python
data2 = pd.concat([data1, datakl,target], axis=1)
```

```python
data2.columns
```

```python
from sklearn.utils import shuffle
data3 = shuffle(data2, random_state = 34)
```

```python
num=data3.shape[1]
length=data3.shape[0]
cut=length//10*9
y=data3['target']
del data3['target']
X=data3
y.tail()
```

```python
from sklearn.feature_selection import RFE
#X, y = make_friedman1(n_samples=50, n_features=10, random_state=0)
estimator =LogisticRegression()
selector = RFE(estimator, 11, step=11)
selector = selector.fit(X, y)
```

```python
summary = np.zeros(sum(selector.support_)).tolist()
j=0
k=0
for i in selector.support_:
    j=j+1
    if i==True:
        summary[k]=j-1
        k=k+1
summary
```

```python
X.tail()
```

```python
x=X.iloc[:,summary]
x.tail()
```

```python
X_train, X_test=x.iloc[0:cut],x.iloc[cut:]
y_train, y_test=y.iloc[0:cut],y.iloc[cut:]
print(X_train.size)#通过输出训练集测试集的大小来判断数据格式正确。
print(X_test.size)
print(y_train.size)
print(y_test.size) 
```

```python
model=LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_train_predict=model.predict(X_train)
print('accuracy:',accuracy_score(y_train_predict,y_train))
print('precision:',precision_score(y_train_predict,y_train))
print('recall:',recall_score(y_train_predict,y_train))
print('f1:',f1_score(y_train_predict,y_train))
```

```python
y_test_predict=model.predict(X_test)
print('accuracy:',accuracy_score(y_test_predict,y_test))
print('precision:',precision_score(y_test_predict,y_test))
print('recall:',recall_score(y_test_predict,y_test))
print('f1:',f1_score(y_test_predict,y_test))
```
